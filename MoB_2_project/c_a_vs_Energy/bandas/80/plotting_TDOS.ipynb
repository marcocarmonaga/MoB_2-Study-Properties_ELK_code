```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '/home/marcocarmonaga/Documents/Elk_projects/MoB_2_project/c_a_vs_Energy/bandas/80/TDOS.OUT'
tdos = np.loadtxt(file)
df = pd.DataFrame(tdos, columns=['Energy','DOS'])
```

```python
df['Energy'] = df['Energy'].apply(lambda x: x * 27.2)
df['DOS'] = df['DOS'].apply(lambda x: x * (1 / 27.2))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10,3.3)
plt.xlim(-10, 10)
plt.ylim(df['DOS'].min(),df['DOS'].max())
ax.plot(df['Energy'], df['DOS'], c='black')
ax.plot([0,0],[df['DOS'].min(),df['DOS'].max()], '--', c='gray')
ax.set_xlabel('Energy (eV)')
ax.set_ylabel('Estados / eV')
ax.set_title('80 GPa')
plt.savefig('TDOS.png')
plt.show()
```
